Nombres: Ana, Luis, Carlos, Maria, Pedro

Edad: 23, 25, 22, 24, 26

Ciudad: Quito, Guayaquil, Cuenca, Manta, Quito

Salario:1200, 1500, 1100, 1300, 1400


Devolver:

Empleados mayores a 23 a単os

Empleados de Quito con salario superior a 1200

Calcular el salario anual del empleado

En que ciudad es el salario mejor pagado

```python
import pandas as pd

df = pd.DataFrame({
    'Nombres': ['Ana', 'Luis', 'Carlos', 'Maria', 'Pedro'],
    'Edad': [23, 25, 22, 24, 26],
    'Ciudad': ['Quito', 'Guayaquil', 'Cuenca', 'Manta', 'Quito'],
    'Salario': [1200, 1500, 1100, 1300, 1400]
})
```

```python
#Aplicar el 10% de descuento usan apply y lambda
df['Salario'] = df['Salario'].apply(lambda x: x * 0.9)
display(df)
```

```python
#Salarios menores a $1000 sean bajos y que los demas sean altos
df['Descuento'] = df['Salario'].apply(lambda x: 'Barato' if x < 1000 else 'Caro')
display(df)
```

```python
df['Estado'] = ['ACTIVO', 'inactivo', 'activo', 'ACTIVO', 'inactivo']


df['Estado'] = df['Estado'].replace({
    'ACTIVO':'Activo',
    'activo':'Activo'
})
display(df)
```

```python
Salarios = df.where(df['Salario']> 1200, 'No aplica' )
Salarios
```

```python
#Eliminar valores nulos
# df.dropna()  #Eliminar filas con algun nulo - uncomment if needed
# df.dropna(subset={'Salario'}) #Eliminar solo if Salario is null - uncomment if needed

#Rellenar valores nulos
# df.fillna('Desconocido') # - uncomment if needed
df['Salario'] = df['Salario'].fillna(df['Salario'].mean())   #Rellenar valores nulos - uncomment if needed
display(df)
```
